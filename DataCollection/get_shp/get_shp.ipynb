```python
import requests
import json
import re
import os
import geopandas
```

```python
country_url = 'https://geo.datav.aliyun.com/areas_v3/bound/geojson?code=100000_full'
```

```python
def get_data(url):
    r = requests.get(url)
    data = r.json()['features']
    return data
```

```python
## 省级数据
output = {"type": "FeatureCollection",
             "features":[]}
with open('省级行政区.json','w') as f:   
    country_data = get_data(country_url)
    output['features'].extend(country_data)
    json.dump(output,f,indent=4)
```

```python
def saveShapefile(file_name,output_shapefile_name):
    try:
        data = geopandas.read_file(file_name) 
        localPath = str(output_shapefile_name)
        data.to_file(localPath, driver='ESRI Shapefile', encoding='gbk') 
        print("转化成功,文件存放位置："+localPath)
    except Exception as e:
        print(e)
# saveShapefile('省级行政区.json','省级行政区')
```

```python
## 县级数据
with open('县级行政区.json','w') as f:   
    country_data = get_data(country_url)
    output = {"type": "FeatureCollection","features":[]}
    for province in country_data:
        province_code = province['properties']['adcode']
        pro_url = re.sub('\d+\d',str(province_code),country_url)
        print(province_code,province['properties']['name'],pro_url)            
        
        if province_code == 710000 or province_code == 810000 or province_code == 820000 or province_code == '100000_JD': 
            continue
        else:
            province_data = get_data(pro_url)
            for city in province_data:
                ## 市数据
                city_code = city['properties']['adcode']
                try:
                    city_url = re.sub('\d+\d',str(city_code),pro_url)
                    city_data = get_data(city_url)
                except:
                    city_url = 'https://geo.datav.aliyun.com/areas_v3/bound/geojson?code=' + str(city_code)
                    city_data = get_data(city_url)

                output['features'].extend(city_data)
                print(city['properties']['name']+'下所有区县保存完毕')

    json.dump(output,f,indent=4)
    
    
```

```python
saveShapefile('县级行政区.json','县级行政区')
```
